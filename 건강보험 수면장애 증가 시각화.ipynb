```python
data = pd.read_csv('국민건강보험공단_수면장애(G47) 및 비기질성수면장애(F51) 진료인원_20221231.csv', encoding= 'EUC-KR')
```

```python
data['진료인원(명)'] = data['진료인원(명)'].replace('NaN', np.nan)
data['진료인원(명)'] = data['진료인원(명)'].str.replace(',', '')
data['진료인원(명)'] = data['진료인원(명)'].astype('float64')
```

```python
sleep  = round(data[['진료년도','진료인원(명)']].groupby(by='진료년도').sum()/10000,1)
```

```python
sleep  = sleep.rename(columns = {'진료인원(명)':'진료인원(만 명)'})
```

```python
# 한글 폰트 적용
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
```

```python
sleep.plot(kind='line', title = '국민건강보험공단 수면장애 및 비기질성수면장애 진료인원',figsize=(12,7),colormap = 'Pastel1')
```
